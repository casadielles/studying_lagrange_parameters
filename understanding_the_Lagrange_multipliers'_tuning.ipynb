
# Understanding the Lagrange multipliers' tuning

<br>
<br>

$\begin{equation}
\begin{bmatrix} * & - & + & x\\
                0 & * & - & +\\
                0 & 0 & * & -\\
                0 & 0 & 0 & *
\end{bmatrix}
\end{equation}$

<br>

The strength or weakness of interactions will be determined by the amount of contributions, i.e., the number of the matrix elements involved, i.e., the number of pair of variables correlated.    
<br>

For example, *in the case of low dimensional problems* (small matrix dimension), interaction including further neighbours ($+$ or $x$) are weaker than first neighbours interactions ($-$ or $*$).
<br>
<br>
However, real problems deal with a huge amount of variables (high dimensional matrix ) so that the dimensional effect is negleted. In these cases, the intensity of an interaction can be estimated in function of the number of variables it consists of. For example, an interaction such as $(\sum_{i}x_{i}-c)^2$  will include $*, -, +$ and $x$ elements of the matrix, while only $ - $ elements participate in an interaction like $\sum_{i}x_{i}x_{i+1}=0$.
<br>
<br>
Summarizing, the strength of an interaction of your model can be infered by having some knowledge about how many pairs of variables will be part of the correlation. Then, you can modify the strength by tuning the Lagrange multipliers assigned to the interaction. For example, it makes sense that the interaction $\sum_{i}x_{i}x_{i+1}=0$ will be multiplied by a bigger Lagrange parameter than the one we use for the interaction $(\sum_{i}x_{i}-c)^2$ in order to satify both conditions when you run the annealing optimization.

<br>
